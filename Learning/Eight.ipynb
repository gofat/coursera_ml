```python
k = 0.01
w1 = 0
w2 = 0
l = len(train_goal)
acc = 0.00001
```

```python
import math
```

```python
def dist(w1, w1_new, w2, w2_new):
    return math.sqrt((w1 - w1_new)**2 + (w2 - w2_new)**2)
```

```python
def gradiend_fall(k, w_1, w_2, l, acc, C=0): 
    for i in range(10000):
        x_1 = 0
        x_2 = 0
        for j in range(l):
            x_1 += train_goal[j]*train_fea[j][0]*(1-1/(1+math.exp(-train_goal[j]*(w_1*train_fea[j][0] + w_2*train_fea[j][1]))))
            x_2 += train_goal[j]*train_fea[j][1]*(1-1/(1+math.exp(-train_goal[j]*(w_1*train_fea[j][0] + w_2*train_fea[j][1]))))

        w_1_new = w_1 + k*(1/float(l))*x_1 - k*C*w_1
        w_2_new = w_2 + k*(1/float(l))*x_2 - k*C*w_2
        
        if (dist(w_1, w_1_new, w_2, w_2_new) < acc):
            print dist(w_1, w_1_new, w_2, w_2_new)
            w_1 = float(w_1_new)
            w_2 = float(w_2_new)
            print i
            break
        else:
            w_1 = float(w_1_new)
            w_2 = float(w_2_new)
    return w_1, w_2
```

```python
w_1, w_2 = gradiend_fall(k, w1, w2, l, acc)
```

```python
print w_1, w_2
```

```python
w_1_reg, w_2_reg = gradiend_fall(k, w1, w2, l, acc, C=10)
```

```python
print w_1_reg, w_2_reg
```
